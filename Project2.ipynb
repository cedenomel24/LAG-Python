```python
dfm_metf = dfm.loc[:, ['encounter_id','patient_nbr','race','gender','age','time_in_hospital', 'metformin', 'metformin-rosiglitazone','metformin-pioglitazone', 'readmitted']]
```

```python
grouped_metf = dfm_metf.groupby('metformin').size()
print(grouped_metf)
```

```python
grouped_metf.plot(kind='bar')
plt.show()
```

```python
grouped_metf_ros = dfm_metf.groupby('metformin-rosiglitazone').size()
print(grouped_metf_ros)
```

```python
grouped_metf_pio = dfm_metf.groupby('metformin-pioglitazone').size()
print(grouped_metf_pio)
```

```python

avg_time_per_metf = dfm_metf.groupby('metformin')['time_in_hospital'].mean()
print(avg_time_per_metf)
```

```python
avg_time_per_metf = avg_time_per_metf.sort_values()
```

```python
avg_time_per_metf.plot(kind='bar', color=['skyblue', 'lightgreen', 'sandybrown', 'lightcoral'],edgecolor='black', linewidth=0.5, width=0.7)
plt.xlabel('Dosage During Encounter')
plt.ylabel('Average Days in Hospital')
plt.title('Average Days in Hospital by Metformin Dosage'.upper(), fontfamily='serif')
plt.xticks(rotation=45, ha='right', fontsize=12)
plt.show()
```

**This analysis investigated the relationship between changes in Metformin Dosage (Up, Down, Steady, No) during an encounter and the average length of Hospital Stay.**

Patients who reported an increase in metformin dosage (Up) had the highest average hospital stay duration (5.36 days), while those with steady or no change in dosage (Steady, No) had lower average stays (4.28 and 4.41 days, respectively).
