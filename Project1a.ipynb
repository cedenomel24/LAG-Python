```python
dshs = ds[ (ds['School Level']=='High school') & (ds['Regents Exam'].str.contains('Common Core')) ]
dshs
```

```python
dshsc=dshs[dshs['School Name'].str.contains('Charter')]
dshsc
```

```python
dshsc['School Name'].value_counts()
```

```python
dsf=dshsc[dshsc['School Name'].str.contains('New Visions Charter')]
dsf
```

```python
dsf['Mean Score'] = pd.to_numeric(dsf['Mean Score'], errors='coerce')
```

```python
gdsf=dsf.groupby('School Name')['Mean Score'].mean()
gdsf
```

```python
dsfg=dsf.groupby(['Year','School Name','Regents Exam'])[['Mean Score']].median()
dsfg
```

```python
# prompt: Using dataframe dsfg: line graph

import altair as alt
alt.Chart(dsfg).mark_line().encode(x=alt.X('Mean Score'))
```

```python
# @title Mean Score

from matplotlib import pyplot as plt
dsfg['Mean Score'].plot(kind='line', figsize=(8, 4), title='Mean Score')
plt.gca().spines[['top', 'right']].set_visible(False)
```
